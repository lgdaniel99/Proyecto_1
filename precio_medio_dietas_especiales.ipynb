### Precio Medio por dieta especial Madrid

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('datos_tripadvisor_madrid_limpio.csv')
```

```python

dietas_especiales = ['Halal', 'Kosher', 'Opciones sin gluten', 'Opciones veganas', 'Opciones vegetarianas']
precios_medios_dieta = {}

for dieta in dietas_especiales:
    seleccion = df[(df[dieta] == True) & (df['precio promedio'].notnull())]
    precios_medios_dieta[dieta] = seleccion['precio promedio'].mean()

#gráfica de barras con estos precios medios.
nombres_dieta = list(precios_medios_dieta.keys())
valores_precio = list(precios_medios_dieta.values())






```

```python
plt.figure(figsize=(10, 5)) 
plt.bar(nombres_dieta, valores_precio, color='skyblue') 
plt.xlabel('Tipo de Dieta') 
plt.ylabel('Precio Promedio') 
plt.title('Precio Promedio por Tipo de Dieta Especial') 
plt.xticks(rotation=45) 
```

```python
# Obtener estadísticas descriptivas de la columna 'Precio Promedio'
estadisticas_descriptivas = df['precio promedio'].describe()

print(estadisticas_descriptivas)
```

```python
# Obtener estadísticas descriptivas de la columna 'Precio promedio' para cada categoría de dieta especial
for dieta in dietas_especiales:
    estadisticas = df[df[dieta] == True]['precio promedio'].describe()
    print(f"Estadísticas para la dieta {dieta}:")
    print(estadisticas)
    print("\n")
```

Los precios promedio para las dietas sin gluten, veganas y vegetarianas varían de 21 a 27, con medias que indican que las opciones sin gluten tienden a ser ligeramente más caras.
Los datos generales muestran un precio promedio ligeramente más bajo de aproximadamente 21, con una mediana de 17, lo que sugiere que hay muchas opciones económicas disponibles.
